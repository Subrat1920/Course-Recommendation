```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('Datasets/online_courses_updated.csv')
df = df.drop(columns=['Unnamed: 0'])
```

```python
df.head()
```

```python
df = df[df['enrollment_numbers']>=40000]
df.head()
```

```python
df.shape
```

```python
top_instructors = df.groupby(['instructor', 'instructor_images'])['rating','feedback_score'].mean().sort_values(by=['rating', 'feedback_score'],ascending=[False, False]).reset_index()
top_instructors
```

```python
instructor_name = 'Sarah Lee'
instructor_with_courses = df[df['instructor']==instructor_name].drop_duplicates(subset=['course_name'])
```

```python
instructor_with_top_rated_courses = instructor_with_courses.groupby(['course_name', 'course_images'])['rating'].mean().sort_values(ascending=False).reset_index()
instructor_with_top_rated_courses
```
